## **LINGUISTIC ANALYSIS OF THE COOKING WEBSITE.**

In this interactive essay, I will go through collecting data about cinamon rolls from a cooking website by using web scraping. Various tools and python modules, including **Beautiful Soup**, **NLTK** and **Matplotlib**, were used for the following analysis.


### Data collection

For the linguistic analysis I have chosen the page devoted to cooking. Below I present the data collection process.

```python
import nltk
import requests
page = requests.get("https://joyfoodsunshine.com/cinnamon-rolls-recipe/")
#print(page.content)
```

At this point, we have our new variable called "page" to which our page value has been assigned. However, the output data is not readable at this stage but we can improve it using Beautiful Soup library. 
Let's move to subsection about data preparation.

### Data preparation

Using the Beautiful soup library, we split the text into paragraphs.

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
#print(soup.prettify())
```

```python
paras = soup.find_all('p')
print(paras)
```

Now we have created a new variable "paras" to which all the scratched paragraphs have been assigned.

```python
only_text = []
for para in paras:
    processed_para = para.get_text()
    processed_para = processed_para.strip()
    if len(processed_para) > 1:
        only_text.append(processed_para)
print(only_text)
```

By using an empty list and appending it with a simple for loop, we made the text appear more readable.

### Data analysis 

```python
tuples = []
for sentence in only_text:
    tokenized = nltk.word_tokenize(sentence)
    tagged = nltk.pos_tag(tokenized)
    for item in tagged:
        tuples.append(item)
print(tuples)
```

At this stage by using the word_tokenize nltk function we create an empty tuple in which we insert tokenized words. Each tuple displayed contains a token and its corresponding part of speech. As we can see, there is a lot of unnecessary punctuation here that we need to remove.

```python
import nltk

text_tokens=[]
text_rp = []
punctuation = [',',':','.',"''",'’','“','”','?','!','%','(',')','>','<','$']
for sentence in only_text:
    tokenized = nltk.word_tokenize(sentence)
    for item in tokenized:
        if item in punctuation:
            pass
        else:
            text_tokens.append(item)
    tagged = nltk.pos_tag(tokenized)
    for item in tagged:
        if item[1] in punctuation:
            pass
        else:
            text_rp.append(item)
print(text_rp) 
```

```python
counter_dict = {}
for item in tuples:
    if item[1] not in counter_dict:
        counter_dict[item[1]] = 1
    else:
        counter_dict[item[1]] += 1

print(counter_dict)
```

With this code we can easily count the occurrences of a given part of speech. 
In addition we can visualize the data as a graph using a **matplotlib**. The result is shown below.

```python
import matplotlib.pyplot as mat
mat.figure(figsize=(20,10))
mat.bar(counter_dict.keys(), counter_dict.values(), color='green')
mat.xlabel('Part of speech', size=20)
mat.ylabel('Data visualization: number of appearances', size=15)
mat.show()
```

As we can see common nouns (NN) most often appear in the text.

Now by using nltk probability we can print the number of times each outcome has occurred. 

```python
from nltk.probability import FreqDist
fdist = FreqDist(text_rp)
print(fdist)
```

```python
fdist.most_common(30)
```

The above output shows 30 the most common words.

```python
from nltk.corpus import stopwords
stopsEnglish = stopwords.words('english')
```

```python
sentenceWords = ['Once', 'the', 'homemade', 'cinnamon', 'rolls', 'are', 'arranged', 'in', 'the', 'baking', 'sheet','cover', 'them', 'with', 'a', 'warm', 'damp', 'towel', 'and', 'let', 'them', 'rise', 'until', 'doubled', 'in', 'size']
```

```python
sentenceWithoutStops = []
for word in sentenceWords:
    if word not in stopsEnglish:
        sentenceWithoutStops.append(word)
print(sentenceWithoutStops)

```

During this step, I removed the stopwords from the example sentence in the text above. I did it on a smaller example to better show the possibilities of this method.
